### Preparing our data
We prepared our data by splitting it into train/test/validations split, you can skip it and head right into the `index.ipynb`

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
import os ,shutil
```

```python
test_filepath_fire = 'Dataset/Testing/fire'
test_filepath_nofire = 'Dataset/Testing/nofire'
train_filepath_fire = 'Dataset/Training and Validation/fire'
train_filepath_nofire = 'Dataset/Training and Validation/nofire'
```

```python
test_fire = [file for file in os.listdir(test_filepath_fire) if file.endswith('jpg')]
test_nofire = [file for file in os.listdir(test_filepath_nofire) if file.endswith('jpg')]

train_fire = [file for file in os.listdir(train_filepath_fire) if file.endswith('jpg')]
train_nofire = [file for file in os.listdir(train_filepath_nofire) if file.endswith('jpg')]
```

```python
print(f'test images with fire are {len(test_fire)} images')
print(f'test images with No fire are {len(test_nofire)} images \n')

print(f'train images with fire are {len(train_fire)} images ')
print(f'train images with No fire are {len(train_nofire)} images ')
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(123)
SEED = 42
```

```python
# splitting into train/validation split
fire_train , fire_val = train_test_split(train_fire , test_size=0.19736842 , random_state=SEED)
```

```python
nofire_train , nofire_val = train_test_split(train_nofire , test_size=0.19736842 , random_state = SEED)
```

```python
print(f'train images with fire are {len(fire_train)} images ')
print(f'train images with No fire are {len(nofire_train)} images \n')

print(f'Validation images with fire are {len(fire_val)} images ')
print(f'Validation images with No fire are {len(nofire_val)} images ')
```

```python
# creating new folder to store our splited data
base = 'data'

train_dir = os.path.join(base, 'train')
val_dir = os.path.join(base, 'validation')
test_dir = os.path.join(base,'test')

os.makedirs(train_dir,exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
```

```python
Train_fire = os.path.join(train_dir, 'fire')
Train_nofire = os.path.join(train_dir ,'nofire')

Test_fire = os.path.join(test_dir, 'fire')
Test_nofire = os.path.join(test_dir , 'nofire')

Validation_fire = os.path.join(val_dir , 'fire')
Validation_nofire = os.path.join(val_dir , 'nofire')

os.makedirs(Train_fire, exist_ok=True)
os.makedirs(Train_nofire, exist_ok=True)

os.makedirs(Test_fire, exist_ok=True)
os.makedirs(Test_nofire, exist_ok=True)

os.makedirs(Validation_fire, exist_ok=True)
os.makedirs(Validation_nofire, exist_ok=True)
```

```python
for image_file in fire_train:
    shutil.copy(os.path.join('Dataset/Training and Validation/fire', image_file), 
                os.path.join(Train_fire, image_file))

for image_file in nofire_train:
    shutil.copy(os.path.join('Dataset/Training and Validation/nofire', image_file), 
                os.path.join(Train_nofire, image_file))

for image_file in fire_val:
    shutil.copy(os.path.join('Dataset/Training and Validation/fire', image_file), 
                os.path.join(Validation_fire, image_file))

for image_file in nofire_val:
    shutil.copy(os.path.join('Dataset/Training and Validation/nofire', image_file), 
                os.path.join(Validation_nofire, image_file))

for image_file in test_fire:
    shutil.copy(os.path.join('Dataset/Testing/fire', image_file), 
                os.path.join(Test_fire, image_file))

for image_file in test_nofire:
    shutil.copy(os.path.join('Dataset/Testing/nofire', image_file), 
                os.path.join(Test_nofire, image_file))
```

```python
test_filepath_fire = 'data/test/fire'
test_filepath_nofire = 'data/test/nofire'

train_filepath_fire = 'data/train/fire'
train_filepath_nofire = 'data/train/nofire'

val_filepath_fire = 'data/validation/fire'
val_filepath_nofire = 'data/validation/nofire'
```

```python
test_fire = [file for file in os.listdir(test_filepath_fire) if file.endswith('jpg')]
test_nofire = [file for file in os.listdir(test_filepath_nofire) if file.endswith('jpg')]

train_fire = [file for file in os.listdir(train_filepath_fire) if file.endswith('jpg')]
train_nofire = [file for file in os.listdir(train_filepath_nofire) if file.endswith('jpg')]

val_fire = [file for file in os.listdir(val_filepath_fire) if file.endswith('jpg')]
val_nofire = [file for file in os.listdir(val_filepath_nofire) if file.endswith('jpg')]
```

```python
print(f'test images with fire are {len(test_fire)} images')
print(f'test images with No fire are {len(test_nofire)} images \n')

print(f'train images with fire are {len(train_fire)} images ')
print(f'train images with No fire are {len(train_nofire)} images ')

print(f'val images with fire are {len(val_fire)} images ')
print(f'val images with No fire are {len(val_nofire)} images ')
```
